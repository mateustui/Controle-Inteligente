**Criação dos dados**

```python
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn


Namostras = 1000
cov = 0.50
X1 = mvn.rvs(mean = np.array([-1, 1]), cov = np.array([[cov, 0], [0, cov]]), size = (Namostras//2))
X2 = mvn.rvs(mean = np.array([1, -1]), cov = np.array([[cov, 0], [0, cov]]), size = (Namostras//2))
X = np.concatenate([X1, X2], axis = 0)
T = np.array([0]*(Namostras//2) + [1]*(Namostras//2))
```

```python
#fig = px.scatter_3d(X[:,0],X[:,1],T, opacity=0.7)
#fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(x=X[:,0], y=X[:,1], z=T,
                                   mode='markers',marker=dict(
        size=1,
        color=T,                # set color to an array/list of desired values
        colorscale='viridis',   # choose a colorscale
        opacity=1
    ))])
fig.show()  
```

```python
mu1=np.mean(X[T==1], axis=0)
mu0=np.mean(X[T==0], axis=0)
cov =np.cov(X[T==1].T)
alpha=len(T[T==0])/(len(T[T==0])+len(T[T==1]))

w=(mu1.T-mu0.T)@np.linalg.inv(cov)
b=(mu0.T@np.linalg.inv(cov)@mu0-mu1.T@np.linalg.inv(cov)@mu1)/2-np.log(alpha/(1-alpha))


def sigmoid(z):
    return(1/(1+np.exp(-z)))
y=sigmoid(X@w+b)
```

```python
##Erro de entropia cruzada

J=-(T*np.log(y)+(1-T)*np.log(1-y)).sum()
print(J)
```
