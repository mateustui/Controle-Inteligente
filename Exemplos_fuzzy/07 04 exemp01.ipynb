```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_ser=np.arange(0,10.1,0.1)
x_com=np.arange(0,10.1,0.1)
x_tip=np.arange(0,25.1,0.1)
```

```python
ser_lo=fuzz.trimf(x_ser,[0,0,5])
ser_me=fuzz.trimf(x_ser,[0,5,10])
ser_hi=fuzz.trimf(x_ser,[5 ,10,10])
```

```python
plt.plot(x_ser,ser_lo)
plt.plot(x_ser,ser_me)
plt.plot(x_ser,ser_hi)
```

```python
com_lo=fuzz.trimf(x_com,[0,0,5])
com_me=fuzz.trimf(x_com,[0,5,10])
com_hi=fuzz.trimf(x_com,[5 ,10,10])
```

```python
plt.plot(x_com,com_lo)
plt.plot(x_com,com_me)
plt.plot(x_com,com_hi)
```

```python
tip_lo=fuzz.trimf(x_tip,[0,0,12.5])
tip_me=fuzz.trimf(x_tip,[0,12.5,25])
tip_hi=fuzz.trimf(x_tip,[012.5 ,25,25])
```

```python
plt.plot(x_tip,tip_lo)
plt.plot(x_tip,tip_me)
plt.plot(x_tip,tip_hi)
```

```python
com_act_lo = fuzz.interp_membership(x_com, com_lo,6.5)
com_act_me = fuzz.interp_membership(x_com, com_me,6.5)
com_act_hi = fuzz.interp_membership(x_com, com_hi,6.5)

print(com_act_lo)
print(com_act_me)
print(com_act_hi)
```

```python
ser_act_lo = fuzz.interp_membership(x_ser, ser_lo,9.8)
ser_act_me = fuzz.interp_membership(x_ser, ser_me,9.8)
ser_act_hi = fuzz.interp_membership(x_ser, ser_hi,9.8)

print(ser_act_lo)
print(ser_act_me)
print(ser_act_hi)
```

```python
regra1_ant=np.fmin(com_act_lo,ser_act_lo)
regra1 = np.fmin(regra1_ant,tip_lo)

regra2 = np.fmin(ser_act_me,tip_me)

regra3_ant=np.fmax(com_act_hi,ser_act_hi)
regra3 = np.fmin(regra3_ant,tip_hi)

agregado = np.fmax(regra3, np.fmax(regra1, regra2))
```

```python
plt.plot(x_tip,agregado)
```

```python
tip =fuzz.defuzz(x_tip,agregado,'centroid')
```
