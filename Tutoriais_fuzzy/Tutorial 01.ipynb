```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10.01, 0.01)
```

```python
print(x)
```

```python
mf1 = fuzz.trimf(x, [2, 5, 9])
```

```python
plt.plot(x, mf1)
```

```python
mf2 = fuzz.trapmf(x, [1, 2, 5, 9])
plt.plot(x, mf2)
```

```python
mf3 = fuzz.gaussmf(x, 5, 1)
plt.plot(x, mf3)
```

```python
mf4 = fuzz.gauss2mf(x, 5, 1.5, 7, 2)
plt.plot(x, mf4)
```

```python
mf5 = fuzz.sigmf(x, 5, +5)
plt.plot(x, mf5)
```
